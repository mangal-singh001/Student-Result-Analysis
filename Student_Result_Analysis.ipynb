Gender Distribution

```python
data['Gender'].value_counts()
```

```python
plt.figure(figsize=(4,4))
ax = sns.countplot(data=data,x='Gender')
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

From the above chart we have analysed that : The number of females is more than the number of males 

```python
gb = data.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between Parent's Education and Student's Score")
plt.show()
```

From the above chart we have concluded that the education of the parent have a good impact on the student marks 

```python
gb1 = data.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Relationship between Parent Marital Status and Student's Score")
plt.show()
```

From the above graph we have concluded there is no / negligible impact on the Student's Score dur to their parent's  marital status

Dsitribution of Ethnic Groups

```python
groupA = data.loc[(data['EthnicGroup'] == 'group A')].count()
groupB = data.loc[(data['EthnicGroup'] == 'group B')].count()
groupC = data.loc[(data['EthnicGroup'] == 'group C')].count()
groupD = data.loc[(data['EthnicGroup'] == 'group D')].count()
groupE = data.loc[(data['EthnicGroup'] == 'group E')].count()

plt.figure(figsize=(10,5))
l= ['group A','groupB','groupC','groupD','groupE']
mlsit = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupA['EthnicGroup']]
plt.pie(mlsit,labels=l,autopct="%.2f%%",startangle=90)
plt.title('Distribution of Ethnic Group')
plt.legend()
plt.show()
```
